```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
name="Navya reddy"
hastags=["#eat","#sleep","#reels","#laugh","#cry","#again sleep"]
text=(name+' ')*100+' '.join(hastags*5)
mask=np.array(Image.open("hmp_img2.webp"))
wordcloud=WordCloud(background_color="white",mask=mask,contour_color="black",contour_width=20,colormap="Dark2").generate(text)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```
